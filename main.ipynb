## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n<=0:
        return 1
    else:
        fact = n*factorial(n-1)
        return fact
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
def unique(lst_un):
    unique = []
    for num in lst_un:
        if num not in unique:
            unique += [num]
    return unique
```

```python
def mode_counter(arr):
#your code here
    max_count = 0
    mode_value = None
    for most_freq in arr:
        count = 0
        for x in arr:
            if x == most_freq:
                count += 1
        if count > max_count:
            max_count = count
            mode_value = most_freq
    return mode_value
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    found=[False]*26
    for letter in string:
        if "A" <= letter <= "Z":
            letter = chr(ord(letter)+32)
        if "a" <= letter <= "z":
            index = ord(letter) - ord ("a")
            found[index] = True
    for element in found:
        if not element:
            return False
    return True
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    words = []
    current_word = ""
    
    for word in string:
        if word == ",":
            words.append(current_word)
            current_word = ""
        else:
            current_word += word

    words.append(current_word)
    words = sorted(words)
    result = ""

    for i in range(len(words)):
        result += words[i]
        if i < len(words) -1:
            result += ","
    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    if not password:
        return False
    
    is_upper = False
    is_lower = False
    is_digit = False
    is_special = False

    special_ele = "!@#$%&*()}{[]-+?_=,></."
    upper_ele = "ABCDEFGHIJKLMNOPQRSTUVYXWZ"
    lower_ele = "abcdefghijklmnopqrstuvyxwz"
    digit_ele = "0123456789"

    len_pass = 0
    for ele in password:
        len_pass +=1

    if len_pass < 8:
        return False
    
    for ele in password:
        found_type = False
        for top_ele in upper_ele:
            if ele == top_ele:
                found_type = True
                is_upper = True
                break
        if not found_type:
            for down_ele in lower_ele:
                if ele == down_ele:
                    found_type = True
                    is_lower = True
                    break
        if not found_type:
            for dig_ele in digit_ele:
                if ele == dig_ele:
                    found_type = True
                    is_digit = True
                    break 
        if not found_type:
            for spe_ele in special_ele:
                if ele == spe_ele:
                    found_type = True
                    is_special = True
                    break
    
    if is_upper and is_lower and is_digit and is_special:
        return True
    return False

```

```python
# This will test your function 
test_pass(check_pass)
```
